### 課題14.1
サイコロを$2^1、2^2、\cdots 2^{10}$回振るシミュレーションを行い、1の目が出る確率を折れ線グラフでプロットしなさい。（X軸：回数、Y軸：確率）

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# フォントのパスを指定
font_path = '/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc'
font_prop = fm.FontProperties(fname=font_path)

# サイコロを振る関数
def roll_dice(num_rolls):
    results = np.random.randint(1, 7, size=num_rolls)
    return np.sum(results == 1) / num_rolls

# 回数リスト (2^1, 2^2, ..., 2^10)
rolls = [2**i for i in range(1, 11)]

# 1の目が出る確率を計算
probabilities = [roll_dice(num) for num in rolls]

# プロット
plt.figure(figsize=(10, 6))
plt.plot(rolls, probabilities, marker='o', linestyle='-', color='b')
plt.xscale('log', base=2)
plt.xlabel('回数（対数スケール）', fontsize=14, fontproperties=font_prop)
plt.ylabel('1の目が出る確率', fontsize=14, fontproperties=font_prop)
plt.title('サイコロを振る回数と1の目が出る確率', fontsize=16, fontproperties=font_prop)
plt.grid(True)
plt.show()
```

### 課題14.2
講義資料の「1Dランダムウォークの統計的性質」という節では，ランダムウォークのサンプルパスを1本生成して，時系列グラフで表示しました．それを踏まえて，サンプルパスの本数を10本生成し，一つの時系列グラフに表示しなさい。このとき各サンプルごとにseedを必ず変更すること（同じ初期値から始めると同じ時系列が生成されることになります）。

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# フォントのパスを指定
font_path = '/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc'
font_prop = fm.FontProperties(fname=font_path)

# 1Dランダムウォークを生成する関数
def generate_random_walk(steps, seed):
    np.random.seed(seed)
    walk = np.cumsum(np.where(np.random.rand(steps) < 0.5, 1, -1))
    return walk

# サンプルパスの本数とステップ数
num_samples = 10
num_steps = 1000

# グラフの描画
plt.figure(figsize=(10, 6))

for i in range(num_samples):
    walk = generate_random_walk(num_steps, seed=i)
    plt.plot(walk, label=f'サンプル {i+1}')

plt.xlabel('ステップ数', fontsize=14, fontproperties=font_prop)
plt.ylabel('位置', fontsize=14, fontproperties=font_prop)
plt.title('1Dランダムウォークのサンプルパス', fontsize=16, fontproperties=font_prop)
plt.legend(loc='upper left', prop=font_prop)
plt.grid(True)
plt.show()
```
