## レポート課題7-2

1から100までの整数の中で，「3で割り切れるか，または10進表記で3を含む」… (※) という条件を満たすものを<u>小さい方から</u>順に保持するリストと，(※) の条件に該当しないものを<u>大きい方から</u>順に保持するリストを作成し，それぞれ `print()` 関数で表示しなさい．（他の人と異なるプログラムを評価します）

```python
satisfy_list = []
not_satisfy_list = []

def check_number(num):
    if num % 3 == 0 or '3' in str(num):
        satisfy_list.append(num)
    else:
        not_satisfy_list.append(num)

for i in range(1, 101):
    check_number(i)

not_satisfy_list = not_satisfy_list[::-1]

print("条件を満たすリスト:", satisfy_list)
print("条件を満たさないリスト:", not_satisfy_list)
```

下記のプログラムコードは，「エラトステネスの篩（ふるい）」のアルゴリズムに基づいて，100 未満の素数を小さい方から並べたリストを作ろうとしたものです．コードに多くのエラー（バグ）があるため，デバッグして正しい結果が得られるようにしてください．Python の文法や機能について分からない点がある場合は，これまでの授業資料を参照するか，自分で調べましょう．

```python
import math

n_max = 100
p_list = list(range(2, n_max))

for n in range(2, int(math.sqrt(n_max)) + 1):
    if n in p_list:
        for multiple in range(n * n, n_max, n):
            if multiple in p_list:
                p_list.remove(multiple)

print(p_list)
```
